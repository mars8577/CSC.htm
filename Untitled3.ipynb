```python
#NUMBERS(INTEGERS AND FLOATS)
#integer operations
orders_monday = 45
orders_tuesday = 52
total_orders = orders_monday + orders_tuesday
print(f"Total orders: {total_orders}")

#Float operatios
price = 29.99
tax_rate = 0.08
tax_amount = price * tax_rate
total_price = price + tax_amount
print(f"Price: ${price}, Tax: ${tax_amount: .2f}, Total: ${total_price: .2f}")
```

```python
#BASIC OPERATIONS
#Real-World calculation : Sales metrics
total_sales = 15000
num_salespeople = 8
commission_rate = 0.05

#Average sales per person
avg_sales = total_sales / num_salespeople
print(f"Average sales per person: ${avg_sales:,.2f}")

#Total commission
total_commission = total_sales * commission_rate
print(f"Total commission: ${total_commission:,.2f}")

#Commission per person
commission_per_person = total_commission / num_salespeople
print(f"commission per person: ${commission_per_person:,.2f}")
```
