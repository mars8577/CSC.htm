String Indexing and Slicing

```python
string = "Python"
negative_index =[-0, -1, -2,-3, -4, -5]
positive_index =[0, 1, 2, 3, 4, 5]

#Data Analysis
# Accessing individual characters
print(f"First character: {string[0]}")
print(f"Second character: {string[1]}")
print(f"Third character: {string[2]}")
print(f"Fourth character: {string[3]}")
print(f"Fifth character: {string[4]}")
print(f"Last character: {string[5]}")

#Slicing: gettin a substring
#synthas: string[start: end] (end is exclusive)

print(f"First 4 characters: {string[0:4]}")
print(f"characters 1-6: {string[5:1]}")
print(f"From start to index 4: {string[:3]}")
print(f"From index 4 to end: {string[5]}")
print(f"Second  to last: {string[-2]}")
print(f"Last 6 character: {string[-3]}")
```
