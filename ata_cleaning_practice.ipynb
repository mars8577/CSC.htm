## Solution

```python
clean_df = df.copy()

# Clean columns
clean_df.columns = clean_df.columns.str.strip().str.lower().str.replace(" ", "_")

# Drop duplicates
clean_df = clean_df.drop_duplicates()

# Clean names
clean_df["name"] = clean_df["name"].str.title()

# Clean major
clean_df["major"] = clean_df["major"].replace("", np.nan).str.strip().str.lower()
major_map = {
    "data science": "Data Science",
    "computer science": "Computer Science",
    "comp sci": "Computer Science",
    "statistics": "Statistics",
    "math": "Mathematics",
    "mathematics": "Mathematics"
}
clean_df["major"] = clean_df["major"].map(major_map)

# Convert numeric
clean_df["age"] = pd.to_numeric(clean_df["age"], errors="coerce")
clean_df["hours_studied"] = pd.to_numeric(clean_df["hours_studied"], errors="coerce")
clean_df["exam_score"] = pd.to_numeric(clean_df["exam_score"], errors="coerce")

# Clean submitted_assignment
clean_df["submitted_assignment"] = clean_df["submitted_assignment"].replace("", np.nan).str.strip().str.lower()
assignment_map = {"yes": "Yes", "y": "Yes", "no": "No", "n": "No"}
clean_df["submitted_assignment"] = clean_df["submitted_assignment"].map(assignment_map)

# Convert dates
clean_df["date_submitted"] = pd.to_datetime(clean_df["date_submitted"], errors="coerce", format="mixed")

# Handle missing
clean_df = clean_df.dropna(subset=["exam_score"])
clean_df["hours_studied"] = clean_df["hours_studied"].fillna(clean_df["hours_studied"].median())
clean_df["major"] = clean_df["major"].fillna("Undeclared")

clean_df
```
